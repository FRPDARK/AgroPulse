```python
import requests
import csv
from bs4 import BeautifulSoup
url = "https://www.canalrural.com.br/nacional/parana/"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

titulos = soup.find_all('h2')

with open('titulo.csv', mode = 'w', newline ='', encoding='utf-8' ) as arquivo_csv:

  escritor = csv.writer(arquivo_csv)
  escritor.writerow(['titulo'])

  for titulo in titulos:
    texto = titulo.get_text(strip = True)
    escritor.writerow([texto])
    print(texto)
```

```python
  import requests
  import selenium
  from bs4 import BeautifulSoup
  import time

  TOKEN = '7828762741:AAFrcjAm0lX2qPXHNhDgRMqkoTghb6MlzCo'
  URL = f'https://api.telegram.org/bot{TOKEN}/'

  def get_updates(offset=None):
      params = {'timeout': 100, 'offset': offset}
      resp = requests.get(URL + 'getUpdates', params=params)
      try:
          return resp.json()
      except:
          return {}

  def send_message(chat_id, text):
      params = {'chat_id': chat_id, 'text': text, 'parse_mode': 'Markdown'}
      requests.post(URL + 'sendMessage', data=params)

  #notÃ­cias
  def get_noticias():
      url = "https://www.canalrural.com.br/nacional/parana/"
      response = requests.get(url)
      soup = BeautifulSoup(response.text, 'html.parser')
      titulos = [h2.get_text(strip=True) for h2 in soup.find_all('h2')]
      msg = "*NotÃ­cias mais recentes:*\n\n"
      for i, t in enumerate(titulos[:5], 1):
          msg += f"{i}. {t}\n"
      return msg

  #clima
  def get_clima():
    url = "https://www.agrolink.com.br/agrotempo/pr/guarapuava/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    blocos = soup.select('div.block-forecast')
    msg = "* Tempo em Guarapuava:*\n\n"

    if not blocos:
        return msg + "_NÃ£o foi possÃ­vel obter os dados do tempo._"

    for bloco in blocos[:2]:
        linhas = bloco.get_text(separator='\n', strip=True).split('\n')
        for linha in linhas:
            msg += f"â€¢ {linha}\n"
        msg += "\n"

    return msg


  # cotaÃ§Ãµes
  def get_cotacoes():
    url = "https://agro.estadao.com.br/cotacoes/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    blocos = soup.select('div.container.flex.mx-auto.py-4')

    msg = "*CotaÃ§Ãµes agrÃ­colas:*\n\n"

    for bloco in blocos:
        titulo = bloco.find('h2')
        corpo = bloco.find_all('p')

        categoria = titulo.get_text(strip=True) if titulo else "CotaÃ§Ãµes"
        msg += f"*{categoria.upper()}*\n"
        msg += "-" * 30 + "\n"

        for linha in corpo:
            texto = linha.get_text(strip=True)
            if texto:
                msg += f"â€¢ {texto}\n"

        msg += "\n"

    return msg


  #resposta
  def main():
      update_id = None
      print("Bot escutando... envie uma mensagem como 'noticias', 'clima' ou 'cotacao'.")

      while True:
          updates = get_updates(offset=update_id)
          if 'result' not in updates:
              continue

          for update in updates['result']:
              update_id = update['update_id'] + 1
              if 'message' in update:
                  chat_id = update['message']['chat']['id']
                  text = update['message'].get('text', '').lower()

                  print(f"Recebido: {text} de {chat_id}")

                  if "noticia" in text:
                      send_message(chat_id, get_noticias())
                  elif "clima" in text:
                      send_message(chat_id, get_clima())
                  elif "cotacao" in text:
                      send_message(chat_id, get_cotacoes())
                  else:
                      send_message(chat_id, "Envie uma palavra-chave como:\n- *noticias*\n- *clima*\n- *cotacao*")

          time.sleep(1)

  if __name__ == '__main__':
      main()
```
